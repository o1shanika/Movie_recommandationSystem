```python
import pandas as pd
```

```python
data=pd.read_csv('dataset.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data['tags']=data['genre']+data['overview']
```

```python
data.head()
```

```python
new_df=data[['id','title','genre','overview','tags']]
```

```python
new_df=new_df.drop(columns=['genre','overview'])
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features=10000,stop_words='english')
```

```python
vec=cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['title']=='The Godfather']
```

```python
dist=sorted(list(enumerate(sim[0])),reverse=True,key=lambda vec:vec[1])
```

```python
#similarity score
dist
```

```python
for i in dist[0:5]:
   print(new_df.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_df[new_df['title']==movies].index[0]
    dist=sorted(list(enumerate(sim[index])),reverse=True,key=lambda vec:vec[1])
    for i in dist[0:5]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
